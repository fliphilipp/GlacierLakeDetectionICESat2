```python
from IPython.display import display, Image, HTML
import os
import shutil
```

```python
src_dir = 0
dst_dir = -1

position = -1

dirdict = {
    0: 'bad_no_lake',
    1: 'good_lake',
    -1: 'lake_bad_depth',
    99: 'interesting'
}
basedir = '/Users/parndt/jupyterprojects/GLD3_complete/AIS/detection_context/'
from_dir = dirdict[src_dir]
to_dir = dirdict[dst_dir]
from_path = basedir + from_dir
to_path = basedir + to_dir 
filelist = [f for f in os.listdir(from_path) if f.endswith('.jpg')]
filelist.sort()
img_fn = filelist[position]
img_path = from_path + '/' + img_fn
print('\nMoving\n%s\n    from: %s\n    to:   %s\n ' % (img_fn, from_dir, to_dir))
display(Image(img_path, width=800))
```

```python
dest = shutil.move(img_path, to_path + '/' + img_fn)
print('\nmoved file to:\n\n%s\n' % dest)
display(Image(dest, width=250))
```
